# Import Libraries

```python
import numpy as np
import pandas as pd
```

# Data Import

```python
# Import the data
raw_data = pd.read_csv('anime_data_raw.csv')

# make a copy of raw_data and display
data1 = raw_data.copy()
pd.set_option('display.max_columns', None)
data1
```

# Data Preprocessing

```python
data1.describe(include = 'all')
```

```python
# Check for duplicates
data1.duplicated(keep= False).sum()
```

```python
# check the data types of each column
data1.info()
```

```python
# Check for NaN values
data1.isnull().sum()
```

```python
# Remove NaN values based on prediction target for our model 'Rating'
data2 = data1.dropna(subset=['rating'])
data2.reset_index(drop=True, inplace=True)
```

```python
data2.isnull().sum()
```

#### MediaType

```python
# Check NaN values for mediaType
data2[data2['mediaType'].isnull()]
```

```python
# fill Nan values with "Unknown"
data2['mediaType'].fillna("Unknown", inplace=True)
data2['mediaType'].value_counts()
```

#### Duration

```python
# convert 'duration' to minutes
movie_duration = pd.to_timedelta(data2['duration'])
movie_duration = movie_duration.dt.total_seconds()/60
data2['duration_min'] = movie_duration
```

```python
data3 = data2.drop(['duration'], axis=1)
data3.head()
```

```python
# Nan values in 'duration_min'can be inferred from the 'mediaType'
data3['duration_min'] = data3.groupby(['mediaType'])['duration_min'].transform(lambda x: x.fillna(x.median()))
```

```python
data3.isnull().sum()
```

#### Start Year

```python
# check the missing values
data3[data3['startYr'].isnull()]
```

```python
# remove the missing values
data3.dropna(subset=['startYr'],  inplace=True)

# reset the index
data3.reset_index(drop=True, inplace=True)
```

```python
data3.isnull().sum()
```

#### Finish Year

```python
# check the missing values
data3[data3['finishYr'].isnull()]
```

```python
# dataset stops at Year 2020, NaN values can be replaced with 2020
data3['finishYr'].fillna(2020, inplace=True)
```

```python
data3.isnull().sum()
```

#### Season of Release

```python
# The season of release is unknown and cannot be inferred from other columns, it will be filled with unknown

data3['sznOfRelease'].fillna('Unknown', inplace=True)
```

```python
data3.isnull().sum()
```

#### Description

```python
# description column will be dropped as its a long text and for the scope of this analysis will not be used to predict rating

data4 = data3.drop(['description'], axis=1)
```

#### Watched

```python
# NaN values in watched may be derived from the kind of studios that produced the anime, therefore 'studios' will be worked on first.
```

#### Studios

```python
# Observation:
# some studio entries are empty with square brackets
# some anime's are produced by more than 1 studio

# Objective:
# fill the empty entries with "Others"
# Identify the studios that have produced at least 100 animes
```

```python
# filling the empty entries with "Others"
data4['studios'].replace("[]", "Others", inplace=True)
```

```python
# remove the square brackets and quote marks on both ends of the data entries
data4['studios'] = data4['studios'].str.strip("['']")
```

```python
# create a list of all animes'
# some animes' were produced by studio collaborations, it will be separated using ',' as the delimiter for separation.

studios_df = pd.DataFrame(data4['studios'].str.split(",", expand=True).values.flatten(), columns=['studios'])

studios_df
```

```python
# create a list of studios above 100 entries
threshold = 100
val = studios_df['studios'].value_counts()
big_studios = val[val.values >= threshold].index.to_list()
big_studios
```

```python
studio_val = []

# loop through each row in our dataset
for index, row in data4.iterrows():
    # split the comma-separated values in 'studios' to a list but also replace the quotation marks
    studio_list = row['studios'].replace("'","").split(", ")
    # check if any value in the 'big_studios' is in the studio_list
    match = any(value in studio_list for value in big_studios)
    # if there is a match, append the matching value to the studio list
    if match:
        studio_val.append(next(value for value in big_studios if value in studio_list))
    # if there is no match, append 'Unknown' to the studio list
    else:
        studio_val.append("Others")


data4['studio_primary'] = studio_val
```

```python
data4.studio_primary
```

```python
data5 = data4.copy()
```

#### Watched (revisited)

```python
# check the missing values
data5[data5['watched'].isnull()]
```

```python
# Nan values in 'watched' can be inferred from the kind of 'studios'
data5['watched'] = data5.groupby(['studio_primary'])['watched'].transform(lambda x: x.fillna(x.median()))
```

```python
data5.isnull().sum()
```

## Create New Features

```python
# possible new features that can be created from our existing data include;
# years running (finishYr-startYr)
# Studio collaboration (True or False)
```

#### Years running

```python
# find the difference between the finish year and start year
years_running = data5['finishYr'] - data3['startYr']
data5['years_running'] = years_running
```

#### Studio Collaboration

```python
# indicate 1 if there was collaboration, else 0
all_studios = data5['studios'].str.split(", ")
studio_colab = [1 if len(i) > 1 else 0 for i in all_studios]
data5['studio_colab'] = studio_colab
```
